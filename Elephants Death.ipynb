# State-wise Details of Elephant Deaths due to Train Accidents from 2021-22 to 2023-24

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('dataset.csv')

print("Dataset Overview:")
print(data.head())

```

```python
data.describe()
```

```python
data = data.replace('NA', 0).fillna(0)

columns_to_convert = ['2021-22', '2022-23', '2023-24']
data[columns_to_convert] = data[columns_to_convert].astype(int)

# Total deaths per year
total_deaths_per_year = data[columns_to_convert].sum()
print("\nTotal Elephant Deaths per Year:")
print(total_deaths_per_year)

```

```python
# Total deaths per state
data['Total'] = data[columns_to_convert].sum(axis=1)
print("\nTotal Elephant Deaths per State:")
print(data[['State', 'Total']])

```

```python
# Visualizations
# 1. Bar chart of total deaths per year
plt.figure(figsize=(8, 6))
total_deaths_per_year.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Total Elephant Deaths per Year')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

### Observations:
1. The total number of elephant deaths due to train accidents has shown an increasing trend over the three years (2021-22 to 2023-24).
   - In 2021-22, there were approximately 30 deaths.
   - In 2022-23, the number remained consistent at around 30 deaths.
   - In 2023-24, the total increased to 35 deaths, showing a noticeable rise.

2. Despite some variation, the overall trend highlights a growing concern over train-related elephant fatalities.

### Implications:
1. **Rising Threats**: The consistent and rising numbers indicate that train accidents continue to be a significant threat to elephant populations.

2. **Need for Mitigation Measures**: Immediate interventions, such as monitoring train routes in areas with high elephant movement and implementing speed regulations, are necessary to prevent further loss.

3. **Policy Intervention**: The data suggests the need for state and national-level policy changes to safeguard elephant corridors and reduce train-related wildlife accidents.

4. **Conservation Awareness**: These findings underscore the need for increased public and governmental awareness regarding the impact of railway infrastructure on wildlife.

5. **Future Research**: Further studies should explore specific routes and times contributing to higher accident rates to formulate targeted strategies for prevention.


```python
# 2. Bar chart of total deaths per state
plt.figure(figsize=(10, 6))
data.sort_values(by='Total', ascending=False, inplace=True)
plt.bar(data['State'], data['Total'], color='purple')
plt.title('Total Elephant Deaths per State')
plt.xlabel('State')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

### Observations:
#### 1. State-wise Distribution:
- Assam has the highest number of elephant deaths among the states, followed by Odisha and Uttarakhand.
- States such as Karnataka, Tripura, and Uttar Pradesh have the least number of recorded elephant deaths.
#### 2. Cumulative Total:
- The "Total" category highlights the aggregate number of elephant deaths across all states, exceeding 40.
#### 3. Skewed Distribution:
- The majority of deaths are concentrated in a few states, primarily Assam and Odisha, while the other states contribute relatively less to the total number.
#### 4. States with Minimal Contributions:
- States like Kerala, Tripura, and Uttar Pradesh show relatively minimal contributions to the total number of deaths.

### Implications:
#### 1. Conservation Focus:
- States like Assam and Odisha need focused elephant conservation efforts to address high mortality rates. This could involve habitat protection, anti-poaching measures, and mitigating human-elephant conflicts.
#### 2. Policy and Resource Allocation:
- Conservation resources and funding should be disproportionately directed towards states with higher mortality rates, such as Assam and Odisha, to maximize impact.
#### 3. Research Needs:
- Further investigation is needed into the causes of high elephant mortality in states like Assam and Odisha. This could include studying factors such as deforestation, human-wildlife conflicts, and natural calamities.
#### 4. Low Mortality States as Models:
- States with low elephant deaths, such as Karnataka and Kerala, could serve as models for effective conservation strategies that can be replicated in high-risk states.
#### 5. Community Involvement:
- High-mortality states could benefit from greater community involvement and education initiatives to promote coexistence with elephants and reduce conflicts.


```python
# 3. Heatmap of deaths by state and year
import seaborn as sns

plt.figure(figsize=(10, 8))
heatmap_data = data.set_index('State')[columns_to_convert]
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap of Elephant Deaths by State and Year')
plt.xlabel('Year')
plt.ylabel('State')
plt.show()
```

## Observations:
#### 1. Temporal Distribution:
- Elephant deaths show variability across the years 2021-22, 2022-23, and 2023-24.
- The total deaths are highest in the year 2023-24 (17), compared to 15 each in the two preceding years.
#### 2. State-wise Patterns:
- Assam consistently reports the highest number of elephant deaths across all three years, with a slight decline over time (8 in 2021-22, 7 in 2022-23, and 2 in 2023-24).
- Odisha shows an increasing trend in elephant deaths, with the highest in 2023-24 (5 deaths).
- Other states like Uttarakhand and West Bengal exhibit a rise in deaths from 2021-22 to 2023-24.
- States like Tamil Nadu and Karnataka report only sporadic deaths, with no consistent trends.
#### 3. Minimal Contributors:
- States such as Kerala, Tripura, and Uttar Pradesh show very low or no elephant deaths during most years.

## Implications:
#### 1. Hotspot Identification:
- Assam and Odisha remain consistent hotspots for elephant mortality, requiring targeted intervention to understand and mitigate the causes.
#### 2. Emerging Concern in Uttarakhand:
- The noticeable increase in deaths in states like Uttarakhand indicates the need for preventive measures to avoid escalation in future years.
#### 3. Year-specific Trends:
- The spike in deaths in 2023-24 across multiple states warrants further investigation to identify contributing factors like climatic changes, habitat loss, or increased human-animal conflicts.
#### 4. Need for Tailored Strategies:
- While Assam and Odisha require extensive measures to curb persistent high mortality, low-mortality states can focus on maintaining existing conservation success.
#### 5. Research on Root Causes:
- Causes like railway accidents, electrocution, or poaching might differ across states. Conducting region-specific research can provide a clearer picture for formulating mitigation strategies.
