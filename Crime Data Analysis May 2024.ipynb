```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("datafile.csv")  

# Clean column names to remove any leading/trailing spaces
data.columns = data.columns.str.strip()

#  columns for the analysis
trend_data = data[[
    "Heads of Crime", 
    "During the current month", 
    "During the corresponding month of previous year", 
    "During the previous month"
]]

# Aggregate data by "Heads of Crime"
trend_summary = trend_data.groupby("Heads of Crime").sum()

# Plotting trends
plt.figure(figsize=(12, 6))
trend_summary.plot(kind='bar', figsize=(12, 6))

# Add titles and labels
plt.title('Crime Trend Comparison: May 2024 vs Previous Year & Month', fontsize=16)
plt.xlabel('Heads of Crime', fontsize=12)
plt.ylabel('Crime Counts', fontsize=12)
plt.legend(['May 2024', 'May 2023', 'April 2024'], fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Observations:
#### "A - IPC Crime":
This category has the highest number of crimes across all months (May 2024, May 2023, and April 2024).
Crime counts appear relatively stable between May 2023 and May 2024, with a slight variation compared to April 2024.
#### "B - Special & Local Laws":
This category ranks second in crime counts. There is minimal fluctuation between the months under comparison, indicating a steady trend.
#### "D - Crime Against Children":
The crime counts in this category are significantly lower compared to "IPC Crime" and "Special & Local Laws."
Little variation is observed between May 2024, May 2023, and April 2024.
#### "E - Crime Against Scheduled Castes/Tribes by Non-SCs/STs":
This category has the lowest recorded crime counts among all categories.
The values are almost negligible, showing consistency across the time periods.

### Implications:
#### High Priority Areas:
Focus should be given to categories with high crime counts, such as "A - IPC Crime" and "B - Special & Local Laws." These require strategic interventions to address recurring issues.
#### Stability Across Months:
The stability in crime counts across May 2023, May 2024, and April 2024 suggests that crime rates in these categories follow predictable trends.
#### Targeted Interventions:
The relatively low crime counts in "D - Crime Against Children" and "E - Crime Against Scheduled Castes/Tribes" indicate that these areas may already benefit from effective interventions or could require additional focus to further minimize occurrences.
#### Policy Implications:
Policymakers and law enforcement should prioritize resource allocation to high-crime categories while continuing monitoring and preventive measures in low-crime areas.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("datafile.csv") 
data.columns = data.columns.str.strip()

# Aggregating the crime counts by "Heads of Crime"
heads_crime_distribution = data.groupby("Heads of Crime")["During the current month"].sum()

# Aggregating the crime counts by "Major heads"
major_crime_distribution = data.groupby("Major heads")["During the current month"].sum()

# Aggregating the crime counts by "Minor Heads" (handle missing values by dropping them)
minor_crime_distribution = data.dropna(subset=["Minor Heads"]).groupby("Minor Heads")["During the current month"].sum()

# Plotting the distribution of crimes by Heads of Crime (Pie Chart)
plt.figure(figsize=(8, 8))
heads_crime_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap="tab10")
plt.title('Crime Distribution by Heads of Crime (May 2024)', fontsize=14)
plt.ylabel('')  # Remove y-axis label
plt.tight_layout()
plt.show()
```

### Observations:
##### "A - IPC Crime":
- This category accounts for the largest proportion of crimes, representing 60.7% of the total crimes in May 2024.
- It highlights that the majority of reported crimes fall under the Indian Penal Code (IPC).
##### "B - Special & Local Laws":
- This category is the second most significant, contributing 33.4% of the total crimes.
- It suggests that a considerable share of crimes is governed by Special and Local Laws.
##### "D - Crime Against Children":
- This category accounts for a very small proportion of total crimes, at around 5%.
- This indicates that crimes against children are relatively low compared to the other categories.
##### "E - Crime Against Scheduled Castes/Tribes by Non-SCs/STs":
- This category represents the smallest fraction of total crimes, making up only 0.1%.
- It suggests that crimes in this category are relatively rare.

### Implications:
##### Focus Areas:
- The overwhelming proportion of crimes in the IPC Crime and Special & Local Laws categories suggests the need for targeted enforcement and policy interventions in these areas.
- Law enforcement agencies should allocate more resources and training to address the most frequent types of offenses in these categories.
##### Child Protection:
- Although crimes against children make up a smaller share, their sensitivity requires continuous monitoring and dedicated programs to ensure child safety.
##### Minority Protections:
- The low crime rate against Scheduled Castes and Tribes may indicate effective measures in place, but its important to maintain vigilance and ensure these communities feel safe.
##### Policy Impacts:
- The data provides a foundation for decision-makers to design crime prevention strategies, prioritize resources, and monitor the effectiveness of existing laws.


```python
# Sort the distribution by crime counts and take the top 10 categories
top_major_crime_distribution = major_crime_distribution.sort_values(ascending=False).head(10)

# Plotting the top 10 categories with a bar chart
plt.figure(figsize=(10, 6))
top_major_crime_distribution.plot(kind='bar', color='skyblue')
plt.title('Top 10 Major Heads by Crime Distribution (May 2024)', fontsize=14)
plt.xlabel('Major Heads', fontsize=12)
plt.ylabel('Crime Counts', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

- Cheating and kidnapping/abduction cases are comparatively fewer but still notable in the top 10 list.

```python
data['Minor Heads'] = data['Minor Heads'].fillna('Unknown')
# Aggregating crime counts by Minor Heads
minor_head_counts = data.groupby('Minor Heads')['During the current month'].sum().sort_values(ascending=False)

# Plot: Minor Heads
plt.figure(figsize=(12, 8))
minor_head_counts.head(10).plot(kind='barh', color='lightcoral', alpha=0.8)
plt.title('Top 10 Crime Categories by Minor Heads (May 2024)', fontsize=14)
plt.xlabel('Crime Counts', fontsize=12)
plt.ylabel('Minor Heads', fontsize=12)
plt.subplots_adjust(left=0.3)
plt.show()
```

### Observations:
##### Dominance of 'Unknown':
- The category "Unknown" has the highest count, exceeding 4000 cases.
- This indicates a significant portion of data that could not be categorized or might lack clarity in reporting.
##### Prominent Categories:
- "Information Technology Act 2000, 2009" ranks second, reflecting a high number of cyber-related crimes.
- "Simple Hurt" follows, indicating frequent incidents involving minor physical injuries.
##### Traffic and Transportation-Related Crimes:
- Categories such as "National Highways," "Other Roads," "State Highways," and "Of Automobiles - Of Two Wheelers" feature prominently.
- This suggests that crimes related to vehicles or roadways are prevalent.
##### Low Occurrence Categories:
- "Street Gambling (87)," "Other Places," and "Others" have relatively low counts, suggesting these crimes are less frequent in comparison.
  
### Insights and Implications:
##### Data Quality Issues:
- The large number of cases under "Unknown" suggests a need for better classification and reporting mechanisms.
- This might involve training officers to record crimes more accurately or updating the reporting systems.
##### Cyber Crime Concerns:
- The high count for the "Information Technology Act" category highlights the growing prevalence of technology-related offenses.
- Measures like public awareness programs and enhanced digital policing could help curb these crimes.
##### Road Safety and Traffic-Related Offenses:
- Multiple categories related to roads and vehicles indicate recurring issues in these areas.
- Improved road safety measures, enforcement of traffic laws, and targeted campaigns to reduce vehicle-related offenses are necessary.
##### Focus on Minor Categories:
- Although categories like "Street Gambling" and "Other Places" have lower counts, they still contribute to the overall crime landscape and should not be overlooked.
