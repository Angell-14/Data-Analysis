```python
import pandas as pd
file_path = 'rs1.csv'
data = pd.read_csv(file_path)

total_tax = data[['AY 2021-22', 'AY 2022-23', 'AY 2023-24']].sum()

print("Total Tax Collection:\n", total_tax)

```

```python
import matplotlib.pyplot as plt

df = pd.read_csv(file_path)

# Set style for plots
sns.set(style="whitegrid")

# Histogram: Distribution of Tax Collected for Each Year
plt.figure(figsize=(16, 5))
for i, year in enumerate(['AY 2021-22', 'AY 2022-23', 'AY 2023-24']):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[year], bins=10, kde=True, color='skyblue')
    plt.title(f'Distribution of Corporate Tax ({year})')
    plt.xlabel('Tax Collected')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()


```

#### Observations:

##### Highly Skewed Distribution:
The histograms show that the tax collection data is heavily skewed to the right (positive skewness).
Most States/UTs contribute relatively small amounts in corporate taxes, while a few contribute significantly higher amounts, leading to the long tail on the right.

###### Majority States in Low Range:
The majority of States/UTs have tax collections in the lower range (under ₹100,000). This is evident from the large bars clustered on the left.

##### Outliers:
The long tails and the presence of small bars on the far right indicate outliers (States/UTs with exceptionally high tax contributions).

##### Consistent Shape Across Years:
The overall shape of the distribution is similar across the three years. This suggests that the pattern of contributions by States/UTs has remained relatively stable, even though the absolute amounts might have increased.

#### Implications:
There is a significant disparity in corporate tax contributions among States/UTs, with a few dominant contributors likely representing major industrial or financial hubs.
Policies aimed at increasing corporate activities might be focused on regions with lower contributions to help balance the tax base.

```python
# Boxplot: Spread and Outliers of Tax Values
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['AY 2021-22', 'AY 2022-23', 'AY 2023-24']], palette="Set3")
plt.title('Boxplot of Corporate Tax Collections')
plt.ylabel('Tax Collected')
plt.xlabel('Assessment Years')
plt.show()


```

#### Key Insights:

##### Skewness:
The data for corporate tax collections is highly skewed, with most companies paying lower taxes and a few outliers contributing significantly higher amounts.
##### Outliers’ Impact: 
These outliers might have a substantial impact on total tax revenues.
##### Policy Implications:
The large presence of outliers could indicate a need for examining tax structures or compliance measures for high-contributing entities.

```python
# Bar Chart: State-wise Tax Collections for Each Year
df.set_index('States/UT')[['AY 2021-22', 'AY 2022-23', 'AY 2023-24']].plot(
    kind='bar', figsize=(14, 8), stacked=True, colormap='viridis')
plt.title('State-wise Corporate Tax Collections (2021-2024)')
plt.xlabel('States/UT')
plt.ylabel('Tax Collected')
plt.legend(title='Assessment Years')
plt.tight_layout()
plt.show()
```

#### Observations

##### 1. Dominant States:
- Maharashtra:
Consistently the highest contributor, with its collections significantly surpassing all other states/UTs.
Likely due to the presence of numerous corporations, financial hubs, and industries.
- Delhi and Karnataka:
Ranked second and third, with notable contributions in all three years. Likely driven by high corporate presence in Delhi NCR and Bengaluru (tech hub).
##### 2. Minimal Contributors:
Smaller states/UTs like Lakshadweep, Mizoram, and Manipur show minimal tax collections, reflecting their smaller corporate base and economic activity.
##### 3. Foreign Contributions:
"Foreign" category has substantial tax collections, highlighting the importance of multinational corporations operating in India. A noticeable increase in collections over the years, possibly due to stricter taxation policies on foreign entities.
##### 4. Total Collections:
The "Total" bar shows a significant jump across the three years, showcasing overall improvement in corporate tax revenues.
##### 5. State/UT-Specific Observations:
- Gujarat: A steady contributor with visible growth, likely reflecting industrial activity.
- Tamil Nadu and West Bengal: Show moderate but consistent growth over the years.

#### Implications
- Maharashtra dominates corporate tax collections, driven by its financial and industrial sectors.
- Delhi and Karnataka maintain high contributions due to corporate and IT activities.
- Smaller states/UTs contribute minimally, reflecting limited corporate presence.
- Foreign tax collections are significant and growing, highlighting India's global economic integration.
- Total collections show consistent growth, indicating better economic performance and compliance.

```python
# Trends Over Years
# Calculate Growth Rate
df['Growth_21_to_22'] = (df['AY 2022-23'] - df['AY 2021-22']) / df['AY 2021-22'] * 100
df['Growth_22_to_23'] = (df['AY 2023-24'] - df['AY 2022-23']) / df['AY 2022-23'] * 100

# Display the growth rates
print("Yearly Growth Rates:")
print(df[['States/UT', 'Growth_21_to_22', 'Growth_22_to_23']])

# Line Plot: Trends in Tax Collection Over the Years
plt.figure(figsize=(14, 8))
for state in df['States/UT']:
    plt.plot(['AY 2021-22', 'AY 2022-23', 'AY 2023-24'], 
             df[df['States/UT'] == state][['AY 2021-22', 'AY 2022-23', 'AY 2023-24']].values.flatten(),
             marker='o', label=state)

plt.title('Trends in Corporate Tax Collection (State-wise)')
plt.xlabel('Assessment Year')
plt.ylabel('Tax Collected')
plt.legend(loc='upper left', ncol=2, fontsize='small', title='States/UT')
plt.tight_layout()
plt.show()
```

### Analysis of Yearly Growth Rates in Corporate Tax Collections

#### Observations
##### Top Performers in Growth (2021-22 to 2022-23):
1. Ladakh (1750%): Exceptionally high growth, indicating a sudden increase in corporate activity or improved tax enforcement in the region.
2. Mizoram (855.95%): Significant growth, though this dropped drastically in the following year.
3. Nagaland (98.57%) and Meghalaya (89.77%): Demonstrated robust growth, likely due to development projects or improved tax policies.

##### Consistent Growth Leaders:
1. Meghalaya (75.40% in 2022-23) and Puducherry (55.36%) continued their upward trends, indicating expanding corporate bases or sustained compliance efforts.
2. Jammu & Kashmir (44.40%) and Jharkhand (43.34%) showed sustained positive growth, likely due to infrastructural improvements.

##### Declining or Negative Growth States:
1. Arunachal Pradesh (-46.87%) and Mizoram (-92.90%) experienced sharp declines in 2022-23, reflecting reduced corporate activity or tax anomalies.
2. Punjab (-5.76%) and Himachal Pradesh (-6.95%) faced marginal declines, indicating stagnation in corporate tax growth.
   
##### Key Contributors:
1. Maharashtra (13.77%), while not the highest in growth, remains the largest contributor due to its massive tax base.
2. Delhi (14.16%) and Karnataka (23.63%) showed consistent growth, driven by urban economic centers.

##### Foreign Contributions:
Growth in foreign corporate taxes remained strong at 34.51%, emphasizing India's global integration.

##### National Total Growth:
- 34.67% (2021-22) and 17.10% (2022-23): These rates indicate overall economic recovery post-COVID and improving compliance measures. However, the slower growth in the second year suggests stabilization.

#### Implications

##### Regional Development Focus:
States like Ladakh, Meghalaya, and Nagaland show promising growth rates, indicating potential hubs for corporate activity with targeted infrastructure and policy support.

##### Economic Stability in Large States:
States like Maharashtra, Delhi, and Karnataka continue to lead in collections, reinforcing their role as economic powerhouses. Efforts should focus on maintaining their competitiveness.

##### Addressing Declines:
States with negative growth (e.g., Arunachal Pradesh, Mizoram) require investigation to address underlying issues, whether economic, administrative, or compliance-related.

##### Global Integration:
The significant growth in foreign tax collections reflects a positive trend of multinational engagement, necessitating consistent global taxation policies.

##### Strategic Policy Needs:
Regions like Punjab and Himachal Pradesh need tailored strategies to revive corporate growth, such as tax incentives or ease of doing business.

##### Focus on Sustained Growth:
With total growth rates stabilizing, national policies must balance regional equity with sustained economic expansion across all states and UTs.
