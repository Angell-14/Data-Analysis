## 2. Numerical Data

### Visualizations

#### Visualization techniques are effective for identifying patterns and anomalies:

- Histogram: Displays the frequency distribution of numerical data.
- Box Plot: Shows data spread, median, and potential outliers.
- Distplot: Illustrates the combination of histogram and KDE plot.
- Violin Plot: Combines box plot and KDE to show data spread and density.
- Line Plot: Useful for continuous data trends (e.g., time-series data).

#### a. Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'], bins = 5)
```

#### b. Distplot

```python
sns.distplot(df['Age'])
```

#### c. Boxplot

```python
import seaborn as sns
import pandas as pd
df = pd.read_csv('train.csv')
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```

#### d. Violin plot

```python
sns.violinplot(x='Pclass', y='Age', data=df)
plt.title('Violin Plot of Age by Pclass')
plt.show()
```

#### e. Line Plot

```python
# Calculate average age by pclass
avg_age_by_pclass = df.groupby('Pclass')['Age'].mean().reset_index()

# Line plot using Seaborn
sns.lineplot(x='Pclass', y='Age', data=avg_age_by_pclass)
plt.title('Average Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Age')
plt.show()
```

### Performing Descriptive Statistics on the data

#### These methods provide a summary of the numerical data:

- Mean, Median, and Mode: Measure the central tendency of the data.
- Variance and Standard Deviation: Indicate data spread or dispersion.
- Minimum, Maximum, and Range: Highlight the spread and extremes.
- Percentiles and Quartiles: Provide insights into the data distribution (e.g., 25th, 50th, 75th percentiles).
- Skewness and Kurtosis: Assess the asymmetry and the peakedness of the distribution.


```python
df['Age'].mode() #Mode of the age
```

```python
df['Age'].median() #Median of the age
```

```python
df['Age'].mean() #Mean of the age
```

```python
df['Age'].min() #To get the minimum age of the passengers
```

```python
df['Age'].max() #To get the maximum age of the passengers
```

```python
range_value = df['Age'].max() - df['Age'].min()
print(f"Range: {range_value}")
```

```python
# Calculate specific percentiles
percentile_25 = df['Age'].quantile(0.25)  # 25th Percentile (First Quartile)
percentile_50 = df['Age'].quantile(0.50)  # 50th Percentile (Median)
percentile_75 = df['Age'].quantile(0.75)  # 75th Percentile (Third Quartile)

print(f"25th Percentile (Q1): {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile (Q3): {percentile_75}")
```

```python
#visualizing the percentile and Quartile
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for Age
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()
```

```python
df['Age'].skew() #skewness of the data
```

```python
df['Age'].kurtosis #kurtosis of the data
```
